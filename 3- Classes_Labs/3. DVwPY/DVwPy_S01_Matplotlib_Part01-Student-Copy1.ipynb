matplotlib.pyplot is a collection of functions that make matplotlib work like MATLAB. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc.

In matplotlib.pyplot various states are preserved across function calls, so that it keeps track of things like the current figure and plotting area, and the plotting functions are directed to the current axes (please note that "axes" here and in most places in the documentation refers to the axes part of a figure and not the strict mathematical term for more than one axis).

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">LINEWIDTH & LINESTYLES & MARKERS & MARKERSIZE & COLOR</p>

### <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:LEFT; border-radius:10px 10px;">Linewidth</p>

<a href="#toc" class="btn btn-primary btn-sm" role="button" aria-pressed="true" 
style="color:blue; background-color:#dfa8e4" data-toggle="popover">Content</a>

```python
x = np.arange(0, 10)
x
```

```python
fig, ax = plt.subplots(figsize=(12, 6))   #OOP

fig.suptitle("To change the line width, we can use the `linewidth` or `lw` keyword argument", fontsize=18)   #suptitel p ile yazdik, sub ile karismasin diye

ax.plot(x, x-1, color="r", linewidth=0.25) #4 tane plot yapmisim, ayni grafik icinde 4 tane lineplot gelecek
ax.plot(x, x-2, color="blue", lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color="b", lw=10)       #en kalini bu,renkleri ile de linewidth ile de oynayabiliriz

plt.show()
```

```python
x  =np.arange(1, 10)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

ax[0].plot(x, x**2, x, x**3, x, x**4, 'r')

ax[1].plot(x, x**2, x, x**3, x, x**4, 'r')
ax[1].set_xlim([1, 3])
ax[1].set_ylim([1, 50])



plt.show()
```

```python
x  =np.arange(1, 10)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

ax[0].plot(x, x**2, x, x**3, x, x**4, 'r')

ax[1].plot(x, x**2, x, x**3, x, x**4, 'r')
ax[1].set_xlim([1, 3])
ax[1].set_ylim([1, 50])

ax[0].axvline(x=5, ls="--")   #vertical 5 cizgisinden gececek
ax[1].axhline(y=15, ls="--")   #horizantal 15 cizgisinden gececek

plt.show()
```

```python
x = np.arange(1, 10)           #hem 0 dakine hem de 1 dekine v, h ekledik
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
ax[0].plot(x, x**2, x, x**3, x, x**4, 'r')
ax[0].axvline(x=2, ls="--")
ax[0].axhline(y=20, ls="--")
ax[1].plot(x, x**2, x, x**3, x, x**4, 'r')
ax[1].set_xlim([1, 3])
ax[1].set_ylim([1, 50])
ax[1].axvline(x=2, ls="--")
ax[1].axhline(y=20, ls="--")
plt.show()
```
